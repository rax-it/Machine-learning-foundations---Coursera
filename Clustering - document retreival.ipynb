## Compute TF-IDF 

```python
people['word_count'] = gl.text_analytics.count_words(people['text'])
```

```python
people.head()
```

```python
tfidf = gl.text_analytics.tf_idf(people['word_count'])
```

```python
gl.canvas.set_target('ipynb')
```

```python
tfidf.head()
```

```python
people['tfidf'] = tfidf
```

```python
people
```

## Examine TF-IDF for obama article

```python
obama = people[people['name'] == 'Barack Obama']
```

## Nearest neighbor model for document retrieval

```python
knn_model = gl.nearest_neighbors.create(people, features=['tfidf'], label='name')
```

```python
knn_model
```

```python
knn_model.query(obama)
```
